#  خوارزمية A* 

هي خوارزمية تحسب المسافة الأقصر من نقطة الى نقطة أخرى آخذة بعين الإعتبار الحواجز في طريقها.وتعتبر من الخوارزميات المفضلة لحل المشاكل التي يمكن التنبؤ الوصول إليها وتوقع تكلفتها قبل الوصول  

تتجنب المرور بالطرق التي تمتلك تكلفة عالية مما يزيد كفاءتها في حل المشاكل , كما أنها تضمن الوصول إلى الحل اذا كان عدد تفرعات الشجرة التي تتمثل فيها المشكله منتهي , تعتبر أسرع  الخوارزميات التي تعتمد على التوقع في إيجاد الهدف عند إعطاء جميع الخوارزميات قيمة متساوية للتوقع والهدف واحد.

تعتمد سرعتها على جودة التوقع للوصول الى الهدف، حيث اذا قل التوقع الى صفر فإن الخوارزمية تعمل كخوارزمية البحث عن أقصر طريق بدون توقع ، وإذا كان التوقع في افضل حالاته تصبح اسرع الى درجة التخلي عن البحث في شجرة المشكلة وتتخذ فقط في المسار الصحيح إلى أن تصل الهدف .


المساحة المطلوبة من الذاكرة للوصول إلى الهدف = عدد التفرعات لشجرة البحث (يمكن أخذهُ بالمتوسط الحسابي لمختلف التفرعات في كل مستويات شجرة البحث ) مرفوعاً لأس العمق لشجرة البحث (عدد المستويات)


الوقت المطلوب للوصول إلى الهدف = عدد التفرعات لشجرة البحث (يمكن أخذهُ بالمتوسط الحسابي لمختلف التفرعات في كل مستويات شجرة البحث ) مرفوعاً لأس العمق لشجرة البحث (عدد المستويات)


## مكونات خوارزمية A*

- القيمة G  ويعني التكلفة من نقطة البداية الى النقطة الحالية باعتبار الحواجز , تكون في نقطة البداية صفر .
-   القيمة H  ويعني توقع للتكلفة من النقطة الحالية الى نقطة النهاية بدون اعتبار الحواجز , يكون في نقطة النهاية صفر . 
- القيمة F وهي قيمة تساوي مجموع G و H، وتكون قيمة F لنقطة البداية تساوي H بينما تكون لنقطة النهاية تساوي G. 
-  تعتبر جميع النقاط التي يمكن الوصول إليها من نقطة واحدة هم "الأبناء" لتلك النقطة
- والنقطة التي تصل للأبناء تسمى "الأب" 


##   حل المشكلة الموضحة سابقًا باستخدام A* 

يمكن استخدام خوارزمية A* لحل هذه المشكلة باعتبار: 

- قيمة G تمثل طول المسار من حفرة البداية إلى الحفرة اللتي  يتواجد فيها الأرنب حالياً
- قيمة H تمثل طول المسار من الحفرة الحالية إلى حفرة نبتة الجزر بدون اعتبار الحواجز
- قيمة F تمثل مجموع قيمتي G و H عند كل حفرة


## طريقة عمل الخوارزمية

1-	 يتم البحث من نقطة (نقطة البداية)

2-	 بوجود قائمتان:

   -	قائمة تحفظ النقاط التي تمت زيارتها (قائمة الأولى)   
   -	قائمة تحفظ النقاط الجاري اكتشافها (قائمة الثانية)

3-	 يتم وضع نقطة البداية في (قائمة  الثانية)

4-	 يتم تكرار العمليات من الخطوة (5) مادامت تتوفر نقاط في (قائمة الثانية)

5-	 يتم اختيار نقطة من القائمة الثانية بناءً على اقل قيمة (F)  لكل نقطة (في البداية لا يوجد سوى نقطة البداية سيتم اختيارها )

6-	يتم نقل النقطة التي تم اختيارها من الخطوة (5) من القائمة الأولى وحذفها من القائمة الثانية 

7-	يتم اختبار النقطة ما إذا كانت تساوي نقطة الهدف أم لا، واذا كانت ليست نقطة النهاية يتم ادخال ابناء تلك النقطة  للقائمة الثانية اذا لم يكونوا في القائمة الأولى مع تحديد لكل نقطة قيم F وG و H.


## ولزيادة المعرفة عن قيمة التوقع


قيمة التوقع لها خاصيتان:

1-	التوقع المتفائل = حيث تكون قيمة التوقع للوصول الى الهدف اقل من او تساوي التكلفة الحقيقة للوصول الى الهدف . وتكون A*  الأمثل إذا توفرت له قيمة التوقع المتفائل و كان نوع البحث "البحث الشجري"

2-	التوقع المتناسق= حيث تكون قيمة التوقع تجعل قيمة (F) للآباء أقل من (F) للأبناء . وتكونA*  في حالتها الأمثل اذا كانت تستخدم كلاً التوقع المتناسق و إذا كان نوع البحث "البحث في رسم بياني" . 


## طريقة تنفيذ الخوارزمية:

استخدمنا فئة "Vertex" لتمثيل كل حفرة ، حيث كل حفرة تمتلك الخصائص التالية:

- اسم (name) 
- دالة التوقع (h)
- دالة تكلفة المسار (g)
- قائمة من الجيران  (neighbors[])
- الأب (parent)
- دالة تمثل مجموع دالتي التوقع وتكلفة المسار (F)


"IsInNeighborsList" هذه الدالة تستقبل حفرة وتختبر ما إذا كانت تلك الحفرة من الجيران المرتبطين مباشرة مع الحفرة التي استدعت الدالة 

```python

class Vertex:

	def __init__(self, name, h, g=0, neighbors=[], parent=None, f=0 ):
		self.name = name
		self.h=h
		self.g=g        
		self.neighbors=neighbors.copy()        
		self.parent=parent
		self.f=f
        
	def IsInNeighborsList(self, node):
		for x in self.neighbors:
			if node==x.name:
				return True
			return False
```

استخدمنا فئة "Graph" لربط الحفر ببعضها البعض، حيث يمتلك عدة متغيرات:

- "dictionary" وهو نوع من أنواع القوائم حيث استخدمناه لتخزين كائن من نوع "Vertex" ويتم الوصول إليه عن طريق اسم الحفرة
- "edges" هي نوع من القوائم ثنائية الأبعاد يتم فيها تخزين قيمة طول كل مسار من حفرة إلى أخرى
-  "edge_indices" يستخدم للوصول إلى الرقم التسلسلي في "edges" ويمكن العثور على الرقم التسلسلي باستخدام اسم الحفرة


```python

class Graph:
	vertices = {}
	edges = []  
	edge_indices = {} 
	start=None
	goal=None
    
  
	def add_vertex(self, vertex):
		if isinstance(vertex, Vertex) and vertex.name not in self.vertices: 
			self.vertices[vertex.name] = vertex 
			for row in self.edges:
				row.append(0)    
			self.edges.append([0] * (len(self.edges)+1))
			self.edge_indices[vertex.name] = len(self.edge_indices) 
			return True
		else:
			return False


	def add_edge(self, u, v, weight=1):
		if u in self.vertices and v in self.vertices:
			self.edges[self.edge_indices[u]][self.edge_indices[v]] = weight
			self.edges[self.edge_indices[v]][self.edge_indices[u]] = weight
			return True
		else:
			return False

        
	def print_graph(self):
		for v, i in sorted(self.edge_indices.items()):
			print(v + ' ', end='')
			for j in range(len(self.edges)):
				print(self.edges[i][j], end='')
			print(' ')

            
	def get_key(self,val): 
		for key, value in self.edge_indices.items(): 
			if val == value: 
				return key

	def find_neighbors(self,node_name):
		i=0
		while i<len(self.edges):
			if self.edges[self.edge_indices[node_name]][i]>0:
				child_name=self.get_key(i)
				if not self.vertices[node_name].IsInNeighborsList(child_name):             
					self.vertices[node_name].neighbors.append(self.vertices[child_name])
			i=i+1
```

#### خصائص خوارزمية A*:

- "closedList" قائمة تستخدم لتخزين الحفر التي تم زيارتها واكتشافها
- "openList" قائمة تستخدم لتخزين الحفر الجاري اكتشافها
- "path" قائمة تستخدم لتخزين الحفر التي تحقق الحل

#### عمل دالة A*:


- عرفنا طريقة A*  بحيث تستقبل كائن من نوع "Graph" وتحدد نقطة البداية ونقطة الهدف ، يتم حساب F و G لنقطة البداية ثم يتم وضعها في قائمة "open List". بعد كل إضافة في قائمة "open List" يتم ترتيبها عن تصاعدياً بناءً على قيمة F.

- يتم البحث في قائمة "openList" ما دامت تمتلك حفر
- يتم اختيار الحفرة التي تمتلك أقل F كنقطة حالية حتى يتم اكتشافها
- يتم اختبار النقطة الحالية ما إذا كانت هي نقطة الهدف، وإذا كانت كذلك يتم استرجاع قائمة "path"
- إذا لم تكن النقطة الحالية تساوي نقطة الهدف يتم ادخالها في قائمة "closedList" بعد التحقق بعدم وجودها في تلك القائمة وإذا كانت موجودة في قائمة "closedList" يتم العودة لإختيار نقطة حالية جديدة. 
- يتم توليد الحفر الناتجة من الحفرة الحالية ثم يتم إسناد الأب لتلك الحفر بالحفرة الحالية. يتم إيجاد قيمة G لكل حفرة عن طريق الإستعانة بالقائمة ثنائية الأبعاد "edge". ومن ثم يتم حساب التكلفة الحقيقية من حفرة البداية وحتى حفرة الابن 
- ثم يتم حساب قيمة F عن طريق جمع قيمتي G و H لحفرة الابن 
- ثم اضافة حفرة الابن الى قائمة "openList"
- نقطة مهمة:
 - اذا كانت الحفرة في قائمة "closedList" يتم تجاهلها والعودة لاختيار ابن آخر
  

### عمل كل دالة

##### 1- IsInClosedList:
يتم استقبال حفرة ثم يتم اختبار ما إذا كانت في قائمة "closedList" أم لا 

 
##### 2- sortFirst:
في هذه الدالة يتم ترتيب قائمة "openList" 

```python
           
closedList=[]
openList=[]
path=[]           
counter=0 
    
def Astar(graph): 
	start=graph.start
	goal=graph.goal
	start.f=start.h 
	start.g=0        
	openList.append((start.f, start))
	openList.sort(key = sortFirst)      
	while len(openList)>0 : 
		global counter
		counter=counter+1
		current=openList.pop(0)[1]
		if current.name==goal.name:
			closedList.append(current)  
			pathNode=current 
			while pathNode is not None:
				path.append(pathNode.name)
				pathNode=pathNode.parent             
			return path
		if not IsInClosedList(current):          
			closedList.append(current)        
			i=0              
			while len(current.neighbors)>i: 
				if not IsInClosedList(current.neighbors[i]):# هنا يتم إختبار ما إذا كانت الجار الحالي أب للحفرة الحالية أم ابن، فإذا كان أب لا يتم عمل ما بداخل نطاق جملة "if"                
					current.neighbors[i].parent=current
					current.neighbors[i].g=graph.edges[graph.edge_indices[current.name]][graph.edge_indices[current.neighbors[i].name]] #array[0][1]
					current.neighbors[i].g=current.neighbors[i].g+current.g                                                             
					current.neighbors[i].f=current.neighbors[i].g+current.neighbors[i].h
				if IsInClosedList(current.neighbors[i]): 
					i=i+1
					continue
				openList.append((current.neighbors[i].f,current.neighbors[i]))
				openList.sort(key = sortFirst)
				i=i+1
	if len(openList)<=0: 
		return False 

            
     
            
def IsInClosedList(node):
	for x in closedList:
		if node.name==x.name:
			return True



    
def sortFirst(val): 
	return val[0]           
            
       
```

# خوارزمية "simulated annealing"

تعتبر خوارزمية "SA" من الخوارزميات التي لا تضمن الوصول الى الهدف دائما .لكنها تمتلك مميزات منها الحفاظ على مساحة  الذاكرة . 

 وتتمكن من  حل المشاكل التي تمتلك مساحة كبيرة وعدد كبير من النقاط الى الوصول الى الهدف بشكل مبسط من خلال تقليل عدد النقاط التي جاري اكتشافها . 

تستخدم في المشاكل التي لا تهتم لطريقة الوصول وانما تركز على الوصول الى الهدف بأيّ طريق ويمكن ان تقود  الى حل ليس الحل الأمثل لكن حل جيد للمشكلة .  

تعتبر من خوارزميات البحث المحلي وتتمكن من حل مشكله ال القيمة الصغرى المحلية في خوارزمية "Hill climbing",  (القيمة الصغرى المحلية) المقصود بها عدم الوصول الى الهدف بسبب ان الخيارات التي يمكن اختيارها للإنتقال الى نقطة اخرى تنقل الى نقاط أسوء من النقطة الحالية.تسمح بإختيار نقاط سيئة ولكن احتمالية اختيار نقاط سيئة تتناقص تدريجيا تكون الاحتمالية كبيرة  عند النقطة الابتدائية .   

ربما تمتلك مشكلة القيمة الصغرى المحلية لكن بنسبة اقل من خوارزمية "Hill climbing"

لا تبحث في كل الطرق الممكنة لايجاد اقصر مسار , لاتهتم بالمسار  فقط تبحث في مسار واحد لإيجاد الهدف ممايجعها مفضلة لحلول بعض المشاكل التي تاخذ وقت ومساحة اذا تم حلها بخوارزميات التوقع الاعتيادية مثل(A* , greedy,) ..الخ   

## مكونات خوارزمية "simulated annealing"

 - قيمة H وهي قيمة التكلفة للوصول الى الهدف بدون حواجز . 
 - المتغير المحلي وهو النقطة المحلية وتكون في البداية هي نقطة البداية
 - T (temperature) تمثل الحرارة وكلما قلَت الحرارة قلَت امكانية اختيار نقاط سيئة 
 - (T) تكون اعلى قيمة له فالبداية ثم يبدأ بالتناقص تدريجياً , كلما كانت قيمته اكبر ساعدت في الحصول على الهدف 
 - قيمة المقارنة بين التوقع للوصول الى الهدف الحالي والتوقع للنقطة التي سيتم الانتقال لها (E∆)
 -  معادلة الاحتمال وهي تقيس احتمالية السماح باختيار نقاط سيئة 
 -  تعتبر جميع النقاط التي يمكن الوصول إليها من نقطة واحدة هم "الأبناء" لتلك النقطة
 - والنقطة التي تصل للأبناء تسمى "الأب" 


## طريقة عمل الخوارزمية 

1-يتم البحث من نقطة البداية

2- يتم وضع قيمة لحرارة المحدد T

3-يتم تكرار الخطوات من خطوة الرابعة الى انتهاء متغير الحرارة أو ايجاد نقطة الهدف

4- يتم اختبار النقطة الحالية ما إذا كانت نقطة الهدف , إذا كانت كذلك يتم استرجاع المسار للوصول الى الهدف

5-إذا كان اختبار النقطة الحاليه ليس مساوي لنقطة الهدف يتم اختيار النقطة التي سيتم الانتقال لها من ابناء النقطة الحالية 
 بشكل عشوائي

6- يتم اختبار النقطة اللتي تم اختيارها  , هل تمتلك توقع افضل من النقطة الحالية ام لا 
   
   - إذا كانت افضل من النقطة الحالية يتم اختيارها  
   - اذا كانت اسوء من النقطة الحالية يتم اختيارها بناءً على معادلة احتمالية السماح باختيار سيئ 

7- يتم التعديل على متغير الحرارة حيث يقل عن الدورة الاولى وثم تبداء الدورة الثانية

## طريقة تنفيذ الخوارزمية : 

- استخدمنا نفس الهيلكة المستخدمة في خوارزمية "A*" لتمثيل الرسم البياني للمشكلة 
- أولا نحتاج لاستيراد مكتبتين هما "math" و "random".  مكتبة "math" سوف تساعدنا لحساب قيم الدوال الأسية واللوغارتيمية بينما مكتبة "random" سوف تساعدنا لاختيار قيم عشوائية.
- عرفنا ال "SimulatedAnnealing" حيث تستقبل الرسم البياني (graph) وتقوم الدالة بإرجاع حل -في حال وجد-   .
- تعتبر حفرة البداية في اول الخوارزمية هي الحفره الحالية للأرنب 
- يتم اعطاء ال T,k قيم ابتدائية 
- الدالة ستقوم بتكرار الخطوات التالية مادامت قيمة T أكبر من الصفر:
 - يتم اختبار الحفرة الحالية ما إذا كانت حفرة الهدف , إذا كانت كذلك فقد تم ايجاد الهدف ويطبع المسار الذي اوصلني للهدف  
 -  إذا لم تكن الحفرة الحالية هي حفرة الهدف يتم اختبار الحفرة الحالية ما إذا كانت موجودة في قائمة "closedList" ، وإذا كانت كذلك لا يتم إكتشافها. ومن ثم نقوم بتحديث قيمة T عن طريق استدعاء دالة "update_temperature"  
 - إذا كانت الحفرة الحالية ليست موجودة في "closedList" ، يتم إضافتها لتلك القائمة 
 - إذا كانت الحفرة الحالية ليس لديها أية جيران ، يتم التراجع للحفرة السابقة -الأب- لأن حفرة الأب يمكن أن يكون لديه بعض الجيران الذين لم يكتشفوا بعد
 - إذا كانت الحفرة الحالية لديها جيران ، يتم اختيار أحدهم عشوائيًا باستخدام دالة "choice()" الموجودة في فئة "random"
 - إذا كانت حفرة الجار موجودة في "closedList" ، يتم حذفها من قائمة الجيران ثم يتم إعادة اختيار جار جديد
 - يتم حساب قيمة (E∆) ، وهو يساوي الفرق بين قيم التوقع للحفرة الحالية والحفرة التالية "حفرة الجار" 
 - إذا كانت قيمة (E∆) سالبة (مما يعني أن حفرة الجار لها قيمة توقعية أفضل ) يتم جعل الحفرة الحالية تساوي حفرة الجار 
 - إذا كانت قيمة (E∆) سالبة يتم حساب قيمة الاحتمالية
 - القيمة الاحتمالية تساوي exp(-∆E/T) حيث "exp" هي الدالة الأسية ويتم حسابها عن طريق المكتبة المُستدعاة "math"
 - يتم توليد قيمة عشوائية بين الصفر والواحد حتى يتم مقارنتها مع القيمة الاحتمالية. يتم الاستعانة بمكتبة "random" لتوليد هذه القيمة العشوائية 
 - إذا كانت القيمة الاحتمالية أكبر من القيمة العشوائية يتم اختيار حفرة الجار كحفرة حالية ، ثم يتم تحديث كلاً من قيمتي T,k
- في حال وصول قيمة T للصفر والدالة لم تجد حل يتم طباعة "لا يوجد حل ، ربما بسبب مشكلة القيمة الصغى المحلية"

### عمل كل دالة 


#####  1- IsInClosedList:
يتم استقبال حفرة ثم يتم اختبار ما إذا كانت في قائمة "closedList" أم لا 

##### 2-update_temperature:
يتم استقبال قيمة ل T ويتم تحديثها وارجاع القيمة الجديدة 

```python
import random
import math
from math import *


closedList=[]
path=[]         
         
    
    
    
    
def update_temperature(T,k):
	T1=T/log(k+1)
	return T1


def SimulatedAnnealing(graph):
	start=graph.start
	goal=graph.goal    
	current=start
	T=10000**30  
	k=1
	while T>0:
		if current.name==goal.name: 
			pathNode=current 
			print ("تم إيجاد الهدف")
			while pathNode is not None: 
				path.append(pathNode.name)
				pathNode=pathNode.parent
			return path
		if not IsInClosedList(current):         
			closedList.append(current)
			while len(current.neighbors)>0: 
				next_node=random.choice(current.neighbors) 
				if not IsInClosedList(next_node):
					next_node.parent=current
					deltaE=next_node.h-current.h
					if deltaE<0:
						current=next_node                      
					else:
						probability = math.exp(-deltaE / T)
						r=random.random() 
						if probability>r:
							current=next_node                          
					T=update_temperature(T,k)
					k=k+1
					break                    
				if IsInClosedList(next_node):
					current.neighbors.remove(next_node)
			if len(current.neighbors)==0: 
				current.parent.neighbors.remove(current)
				current=current.parent                
		if IsInClosedList(current): 
			T=update_temperature(T,k)
			k=k+1
	return 'لا يوجد حل ، ربما بسبب مشكلة القيمة الصغى المحلية'
        
        
        
def IsInClosedList(node): 
	for x in closedList:
		if node.name==x.name:
			return True
            

            


print(SimulatedAnnealing(graph))
```
